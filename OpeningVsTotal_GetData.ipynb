```python
#Beautiful Soup to get the list of urls for individual movies
url = 'http://www.boxofficemojo.com/yearly/chart/?yr=2015&p=.htm'
response = requests.get(url)
response.status_code
```

```python
page = response.text
```

```python
soup = BeautifulSoup(page, "html.parser")
```

```python
#Get list of URL links to extract individual movie data
a_list = soup.find_all('table')[6].find_all('a', href=True)

id_list = []
k = 9
while k < 309:
    id_list.append(a_list[k])
    k = k + 3

link_list = []
pattern = '"(.*?)"'
for i in id_list:
    m = re.findall(pattern, str(i))
    link_list.append(m)
#for z in link_list:
#    print z
```

```python
df = pd.DataFrame()
```

```python
j = 0
k = 10
#df = pd.DataFrame()
df = []
bad_list = []

while j < len(link_list):
    movie_url = 'http://www.boxofficemojo.com' + str(link_list[j][0])
    holderFrame = pd.read_html(movie_url)
    while k < len(holderFrame) - 1:
        appender = 'N/A'
        genFrame = holderFrame[k]
        if genFrame[0][0] == 'Director:':
            appender = genFrame[1][0]
            break
        if genFrame[0][0] != 'Director:':
            k += 1
    #df.append(genFrame[1][0], ignore_index=True)
    df.append(appender)
    print movie_url, '  ', appender
    k = 10
    j += 1

df  #figure out how to either append to dataframe intially or turn list into dataframe
```
